# Создание виртуальной среды

## Установка пакета
Выполняем установку системного пакета `python3-venv`:
```bash
$ sudo apt-get install python3-venv
```

### Создание виртуальной среды

Инициализируем каталог-виртуальную среду `venv`:
```bash
$ python3 -m venv venv
```

Должно получится что-то вроде этого:
```bash
     $ tree -L 3
     venv
         ├── bin
         │   ├── activate
         │   ├── activate.csh
         │   ├── activate.fish
         │   ├── easy_install
         │   ├── easy_install-3.5
         │   ├── pip
         │   ├── pip3
         │   ├── pip3.5
         │   ├── python -> python3
         │   └── python3 -> /usr/bin/python3
         ├── include
         ├── lib
         │   └── python3.5
         ├── lib64 -> lib
         ├── pyvenv.cfg
         └── share
             └── python-wheels
```

### Активация виртуального окружения
```bash
$ source ./venv/bin/activate
```

## Деактивация виртуального окружения

По окончании работы с виртуальным окружением, его необходимо деактивировать:
```bash
$ deactivate
```

# Jupyter Notebook
В этом приложении мы будем создавать наши ноутбуки.

## Устанавливаем пакет jupyter
```bash
pip install jupyter
```

Вместе с библиотеками установится пакет `ipython`, который является улучшенной версией интерактивного интерпретатора Python c подсветкой синтаксиса, автодополнением через `<TAB>` и другими возможностями. Его можно запустить из консоли командой:
```bash
ipython
```

## Запуск web-приложения jupyter-notebook

Для начала нужно установить пароль на вход web-приложения. Нас попросят ввести пароль `Enter password:` и подтвердить пароль `Verify password:`:
```bash
jupyter-notebook password
```

Затем, можно запустить web-приложение командой `jupyter-notebook`, которая запустит web-сервер и автоматически откроет в браузере страницу http://localhost:8888/:
```bash
jupyter-notebook
```

Справка по запуску web-приложения доступна через команду:
```bash
jupyter-notebook --help
```

Альтернативная команда запуска, где web-приложение будет слушать интерфейс с адресом хоста `pytest`:
```bash
jupyter-notebook --ip=pytest --no-browser
```

## Установка тем jupyter-themes

Проект тем для jupyter доступен по [этому](https://github.com/dunovank/jupyter-themes) адресу.

Выполняем установку:
```bash
pip install jupyterthemes
```

Обновить темы до последней версии можно командой:
```bash
pip install --upgrade jupyterthemes
```
